```python
def count_characters(input_string):
    # Initialize counters
    total_chars = 0
    total_letters = 0
    total_numbers = 0
    total_spaces = 0
    total_vowels = 0
    word_count = 0

    # Create a list of vowels
    vowels = "aeiouAEIOU"

    # Split the input string into words
    words = input_string.split()

    # Iterate through characters in the input string
    for char in input_string:
        total_chars += 1

        # Check if the character is a letter
        if char.isalpha():
            total_letters += 1

        # Check if the character is a digit
        elif char.isdigit():
            total_numbers += 1

        # Check if the character is a space
        elif char.isspace():
            total_spaces += 1

        # Check if the character is a vowel
        if char in vowels:
            total_vowels += 1

    # Calculate the average word length
    if words:
        word_count = len(words)
        total_word_length = sum(len(word) for word in words)
        average_word_length = total_word_length / word_count
    else:
        average_word_length = 0

    return total_chars, total_vowels, average_word_length

def main():
    # Accept input from the user
    user_input = input("Enter a string: ")

    # Count characters, vowels, and calculate average word length
    total_chars, total_vowels, average_word_length = count_characters(user_input)

    # Display the results
    print(f"Total characters (including spaces and numbers): {total_chars}")
    print(f"Total vowels: {total_vowels}")
    print(f"Average word length: {average_word_length:.2f}")

if __name__ == "__main__":
    main()
```
